# Intrdouction to Python - Dictionary - Assignment #5 

## 1.	Create a simple dictionary that stores 2 variables, for example: first and last name.

```
# Initializing an empty dictionary.
simple_dictionary = {}

# Adding key and value elements.
simple_dictionary['first name'] = 'Martin'
simple_dictionary['last name'] = 'Gonzales'
```

```
print(simple_dictionary)
```

## 2.	Print out those variables stored in your previous dictionary. 

```
# Printing using loop.
for keys in simple_dictionary:
  print("variable:\t" + keys)
```

```
# Using other loop style through .items() method.
for keys, values in simple_dictionary.items():
  print(keys)

print('\n')

for keys, values in simple_dictionary.items():
  print(values)
```

## 3.	Add a message to those variables on printing: for example: “Hello, firstname lastname!”

```
print("Hello, " + simple_dictionary['first name'] + " " + simple_dictionary['last name'])
```

```
# Using .format() method for readability (geeksforgeeks.com).
print("Hello, {} {}!".format(simple_dictionary['first name'], simple_dictionary['last name']))
```

## 4.	Create a dictionary that holds 2 key: value pairs:

a.	Look through your dictionary and print each pair,


```
# Initialize and add elements.
new_dict = {}
new_dict['first'] = "Martin"
new_dict['second'] = "John"

print(new_dict)
print(new_dict.items())
```

```
# Printing key and value pairs.
for key, value in new_dict.items():
  print("key: {}\t value: {}".format(key,value))
```

```
# Using one iterator.
for key in new_dict:
  print("key: {}\t value: {}".format(key,new_dict[key]))
```
