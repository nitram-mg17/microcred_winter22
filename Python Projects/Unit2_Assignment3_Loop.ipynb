## 1. Create a list that holds 5 data variables. 

```
# Initializing five variables.
var1 = "name"
var2 = "address"
var3 = "age"
var4 = "id"
var5= "email"
my_list = [var1, var2, var3, var4, var5]
```

```
print(my_list)
```

## 2.	Print out those variables by using a for loop

```
for var in my_list:
  print(var)
```

## 3.	Modify your for loop to add a message to your for loop.

```
for var in my_list:
  print("This variable is named " + var)
```

## 4.	Use a for loop to print even numbers from 1-20.

```
for i in range(2,21,2):
  print(i)
```

## 5.	Sort your list in alphabetic order.

```
# Permanent sorting of list.
my_list.sort()
```

## 7.	Create a loop to print the last 2 elements of your list.

```
# Looping the list backwards using range method.
for i in range(len(my_list)-1, 2, -1):
  print(my_list[i])

print("\n**** or ****\n")

# Looping the list using list slicing.
for i in my_list[-2:]:
  print(i)
```
