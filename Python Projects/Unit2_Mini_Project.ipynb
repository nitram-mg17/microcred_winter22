Ask the user for a number. Depending on whether the number is even or odd, print out an appropriate message to the user. Hint: how does an even / odd number react differently when divided by 2?


```
# Ask for input and typecast into int datatype.
user_number = int(input("Please enter a number: "))
```

```
# Making it a program to keep asking until exit.
```

```
# Exit instruction.
print("Please type e to exit\n")

while True:
  # Ask for input and check for exit flag.
  user_input = input("Please enter a number: ")
  if user_input == "e":
    print("Thank you! Bye!")
    break

  # Type cast into int data type.
  user_number = int(user_input)

  # Odd or even check.
  if user_number % 2 == 0:
    print(user_input + " is even.")
  else:
    print(user_input + " is odd.")
```

# Bonus:

1.	If the number is a multiple of 4, print out a different message.

2.	Ask the user for two numbers: one number to check (call it num) and one number to divide by (check). If check divides evenly into num, tell that to the user. If not, print a different appropriate message.

3. Write a custom function for finding whether the number is odd or even and call it in your main program code.


```
# Updating continuous program with bonus condition 1.

# Exit instruction
print("Please type e to exit\n")

while True:
  # Ask for input and check for exit.
  user_input = input("Please enter a number: ")
  if user_input == "e":
    print("Thank you! Bye!")
    break

  # Type cast into int data type.
  user_number = int(user_input)

  # Check if divisible by 4, then for odd/even.
  if user_number % 4 == 0:
    print(user_input + " is divisible by 4")
  elif user_number % 2 == 0:
    print(user_input + " is even.")
  else:
    print(user_input + " is odd.")
```

```
# Updating continuous program with bonus condition 2.

# Exit instruction
print("Please type e to exit\n")

while True:
  # Ask for input and check for exit flag.
  user_num = input("Please enter a number to check: ")
  user_check = input("Please enter another number to divide by: ")
  if (user_num == "e") or (user_check == "e"):
    print("Thank you! Bye!")
    break

  # Type cast both inputs into int data type.
  num = int(user_num)
  check = int(user_check)

  # Check for divisibility.
  if num % check == 0:
    print(user_check + " divides evenly into " + user_num + ".")
  else:
    print(user_check + " does NOT divide evenly into " + user_num+ ".")
```

```
# Creating odd/even function and calling it for bonus condition 3.
def odd_or_even(number):
  """This function checks for odd/even and prints the result"""
  if number % 2 == 0:
    print(f"{number} is even.")
  else:
    print(f"{number} is odd.")


# Main program.
print("Welcome to my odd or even game!\n\nPlease type e or exit to quit the game\n")

# Exit flag.
keep_going = True

while keep_going:
  # Get input and check for exit condition.
  user_input = input("Please enter a number: ")
  user_input = user_input.lower()
  if user_input == 'e':
    keep_going = False
    print("\nThank you. Bye!")

  else:
    # Call created function to perform ordd/even check.
    user_input = int(user_input)
    odd_or_even(user_input)

  
```
